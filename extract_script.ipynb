```python
from PIL import Image
import requests
from io import BytesIO
from keras.preprocessing import image
import numpy as np
```

```python
file = open("Data/Ski_jumping/images.txt")
tab = file.readlines()
file.close()
cpt = 0
i = 0
while cpt<40:
    url = tab[i]
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        img.save("Data/train_dir/class1/img_c1_"+str(cpt)+".png")
        cpt += 1
        i += 1
        print(cpt)
    except:
        print("Erreur pour "+str(url))
        i += 1

while cpt<50:
    url = tab[i]
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        img.save("Data/val_dir/class1/img_c1_"+str(cpt)+".png")
        cpt += 1
        i += 1
        print(cpt)
    except:
        print("Erreur pour "+str(url))
        i += 1        
```
